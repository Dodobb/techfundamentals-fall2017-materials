```python
#This doesn't actually work in the origional file, as NA's are automatically recoded.  
#But in our file userdata_python.csv it does.  
import numpy as np
#userdata.loc[userdata['avgsales'] == "NA", 'avgsales'] = None
userdata.loc[userdata['avgsales'] == "na", 'avgsales']= None
userdata.head()

#Option2: to get the NAN.
userdatab.loc[userdatab['avgsales'] == "na", 'avgsales']= np.nan
print(userdata.head(), '\n',userdatab.head())
```
